```python
from search import *
from csp import *
from tabulate import tabulate

courses = [f"COURSE{i}" for i in range(1, 5)]
print("Courses:")
print(tabulate([courses], headers="firstrow", tablefmt="grid"))

rooms = [f"Room{i}" for i in range(1, 4)]
print("\nRooms:")
print(tabulate([rooms], headers="firstrow", tablefmt="grid"))

days = ["MW", "TTH"]
periods = ["08:00-09:30",
              #"09:30-11:00",
              #"11:00-12:30",
              #13:30-15:00",
              #"15:00-16:30",
              #"16:30-18:00",
              #"18:00-19:30",
              "19:30-21:00"]
time_slots = [f"{day} - {period}" for day in days for period in periods]
print("\nTime Slots:")
print(tabulate([time_slots], headers="firstrow", tablefmt="grid"))

Domain = [f"{time_slot} - {room}" for time_slot in time_slots for room in rooms]
print("\nDomain:")
print(tabulate([Domain], headers="firstrow", tablefmt="grid"))


#domains = {(course, Room): Domain for course in courses for room in rooms}
domains = {course: Domain for course in courses}
print("\nDomains:")
print(tabulate(domains.items(), headers=["Course", "Domain"], tablefmt="grid"))

neighbors = {course: [other_course for other_course in courses if other_course != course] for course in courses}
print("\nNeighbors:")
print(tabulate(neighbors.items(), headers=["Course", "Neighbors"], tablefmt="grid"))

# Constraint function to ensure no two courses are given in the same time slot
def no_conflict_constraint(course1, time_slot1, course2, time_slot2):
    return (course1 != course2) and (time_slot1 != time_slot2)

# Create an instance of the CSP class
csp_instance = CSP(courses, domains, neighbors, no_conflict_constraint)

"""SOLVE USING tree_csp_solver: """
print("TREE_CSP_SOLVER : ")
solution = tree_csp_solver(csp_instance)
print(solution)

"""SOLVE USING depth_first_graph_search: """
print("\n DEPTH_FIRST_GRAPH_SEARCH : ")
# Solve the CSP problem
csp_solver1 = depth_first_graph_search(csp_instance)
solution_node = csp_solver1
# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")

"""SOLVE USING breadth_first_tree_search: """
print("\n breadth_first_tree_search : ")
csp_solver2 = breadth_first_tree_search(csp_instance)
solution_node = csp_solver2

# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")


"""SOLVE USING depth_first_tree_search: """
print("\n depth_first_tree_search : ")
csp_solver3 = depth_first_tree_search(csp_instance)
solution_node = csp_solver3

# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")


"""SOLVE USING breadth_first_graph_search: """
print("\n breadth_first_graph_search : ")
csp_solver4 = breadth_first_graph_search(csp_instance)
solution_node = csp_solver4

# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")


"""SOLVE USING uniform_cost_search: """
print("\n uniform_cost_search : ")
csp_solver5 = uniform_cost_search(csp_instance, display=False)
solution_node = csp_solver5

# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")


"""SOLVE USING depth_limited_search: """
print("\n depth_limited_search : ")
csp_solver6 = depth_limited_search(csp_instance, limit=50)
solution_node = csp_solver6

# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")


"""SOLVE USING iterative_deepening_search: """
print("\n iterative_deepening_search : ")
csp_solver7 = iterative_deepening_search(csp_instance)
solution_node = csp_solver7

# Print the solution
if solution_node:
    print("Solution found:")
    print(solution_node.state)
else:
    print("No solution found.")


"""SOLVE USING tree_csp_solver: """
print("\n AC3 : ")
solution = AC3(csp_instance)
csp_instance.display(solution)
print(solution)
```
